```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**Read the dataset**

```python
data=np.genfromtxt('c_0000.csv',delimiter=',')
```

```python
data=data[1:64001,:]
```

```python
velx=data[:,3]
```

```python
num_bins=15
plt.figure(figsize=(15,8))
n,bins,patches=plt.hist(velx,num_bins)
plt.grid()
plt.xlabel('x-velocity distribution ')
plt.ylabel('Number of stars')
plt.title(r'Histogram of velocity distributions')
```

```python
vely=data[:,4]
```

```python
plt.figure(figsize=(15,8))
plt.hist([velx,vely],num_bins)
plt.grid()
plt.xlabel('x-velocity distribution ')
plt.ylabel('Number of stars')
plt.title(r'Histogram of velocity distributions')
```
